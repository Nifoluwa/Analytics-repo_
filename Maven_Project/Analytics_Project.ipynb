```python
# Create Sales column
data["sales"] = data["transaction_qty"] * data["unit_price"]
```

```python
locations = data.groupby("store_location").agg({
    "sales":"sum",
    "transaction_id":"count"
    }
)
```

```python
locations
```

```python
plt.pie(
    x=locations["transaction_id"],
    labels=locations.index, 
    shadow=True,
    autopct = "%1.1f%%", 
    explode=[0.2, 0.2, 0.2],)
plt.title("Proportion of visits by store");
#plt.savefig("plot_1.png");
```

```python
plt.pie(
    x=locations["sales"], 
    labels=locations.index, 
    shadow=True, 
    autopct = "%1.1f%%", 
    explode=[0.2, 0.2, 0.2],)
plt.title("Proportion of Revenue($) by store");
#plt.savefig("plot_2.png");
```

```python
data["year"], data["month"], data["day"] = data["transaction_date"].dt.year,data["transaction_date"].dt.month_name(),data["transaction_date"].dt.day_name()
```

```python
data.head()
```

```python
order = ["January", "February", "March", "April", "May", "June"]
```

```python
sales_by_month = data.groupby("month")["sales"].sum().T[order]
```

```python
sales_by_month
```

```python
sales_by_month_individual = data.groupby(["store_location", "month"])["sales"].sum().unstack()[order]
```

```python
sns.lineplot(sales_by_month, marker="o")
plt.annotate(text="81,677$", xy=("January", 83000))
plt.annotate(text="76,145$", xy=("February", 72000))
plt.annotate(text="98,884$", xy=("March", 95000))
plt.annotate(text="118,941$", xy=("April", 115000))
plt.annotate(text="156,727$", xy=("May", 152000))
plt.annotate(text="166,485$", xy=("June", 163000))
plt.xlabel("Month")
plt.ylabel("Total Revenue($)")
plt.title("Total Monthly Revenues", fontsize=20);
#plt.savefig("MR.png");
```

```python
sales_by_month_individual.T.plot(kind="bar", figsize=(10, 7))
plt.tight_layout()
plt.title("Revenues of stores by month($)");
#plt.savefig("plot_4.png");
```
